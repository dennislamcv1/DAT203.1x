```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

import sklearn
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("incomemod.csv")
```

```python
df
```

```python
df = df.drop(['income'],axis=1)
```

```python
df
```

```python
kmeans = KMeans(n_clusters=2,max_iter=500,random_state=4567)
```

```python
df.values
```

```python
kmeans.fit(df.values)
```

```python
df['labels'] = kmeans.labels_
```

```python
df
```

```python
fig, ax = plt.subplots(figsize=(12,9))
sns.scatterplot(x=df.age,y=df.fnlwgt,hue=df.labels,ax=ax)
plt.show()
```

```python
transform = kmeans.fit_transform(df.values)
```

```python
transform
```

```python
transform = pd.DataFrame(transform, columns=['c1','c2'])
```

```python
transform
```

```python
transform['labels'] = kmeans.labels_
```

```python
transform
```

```python
fig, ax = plt.subplots(figsize=(12,9))
sns.scatterplot(x=transform.c1,y=transform.c2,hue=df.labels,ax=ax)
plt.show()
```
